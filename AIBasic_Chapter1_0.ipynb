파이썬 노트북을 편집하는 가장 대표적인 프로그램이 Jupyter Notebook이다. 이 프로그램은 Anaconda 환경을 설치했다면 자동으로 설치된다.  
Jupyter Notebook의 장점은 PC에 설치되어있는 파이썬 환경에서 실행이 가능하다는 것이다. 즉 커스터마이징 부분에서 상당히 유리하다.  
프로그래밍을 많이 하다보면 각자 PC에 자신이 쓰는 언어 관련 셋팅들을 자신이 편한대로 하게 된다. 이 경우에는 Jupyter Notebook이 상당히 편리할 것이다.  
이 수업에서 Colab을 권장하는 이유는 다음과 같다.  
- 대부분의 고딩들이 자신의 컴퓨터 셋팅을 커스터마이징 할 만큼 컴퓨터를 잘 다룬다는 가정을 하고 수업을 했다간 대 참사가 날 수 있다.  
- Jupyter Notebook은 분명 좋은 툴이지만 구동시키는 PC의 파이썬 셋팅 차이가 사람마다 다 다를 수 있다. 
- 그냥 다 각설하고 셋업하기 귀찮으니까. 그리고 컴퓨터에 뭐 설치 안해도 되니까.  

ipynb 즉 파이썬 노트북 파일들은 Jupyter Notebook에서든 Colab에서든 구동 가능하다. 그렇지만 파이썬 코드들을 수정해야 할 경우가 생길 수 있다. (예를 들면 파일 경로 같은 정보들?)
